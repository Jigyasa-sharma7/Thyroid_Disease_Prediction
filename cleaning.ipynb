<h3>making both datasets of same columns first

```python
selected_cols = [
    "age","sex",
    "on_thyroxine","query_on_thyroxine","on_antithyroid_medication",
    "sick","pregnant","thyroid_surgery",
    "query_hypothyroid","query_hyperthyroid",
    "lithium","goitre","tumor",
    "TSH","T3","TT4","T4U","FTI"
]
# These are the columns that we need in our final dataset
```

```python
for col in selected_cols:
    if col not in df_euth.columns:
        df_euth[col] = None
# This will add the columns present in the final selected list which are not in this dataset
```

```python
for col in selected_cols:
    if col not in df_sick.columns:
        df_sick[col] = None
```

```python
df_sick = df_sick[selected_cols]
df_euth = df_euth[selected_cols]
# This will make the datasets in the same order of the selected columns list
```

<h3>Merging of the datasets

```python
# df_sick["target"] = "sick"
# df_euth["target"] = "normal"
# Now we add the target column to both the datasets
df_final = pd.concat([df_sick, df_euth], axis=0)
# We are now merging both datasets vertically
df_final.to_csv("final_thyroid_dataset.csv", index=False)
# Saving the content into new file without extra indexing
```

```python
df_final.shape
```

<h3>Handling missing values

```python
df_final.replace({"t":1, "f":0}, inplace=True) 
# t/f become numeric
```

```python
num_cols = ["TSH","T3","TT4","T4U","FTI","age"]

for col in num_cols:
    df_final[col] = pd.to_numeric(df_final[col], errors="coerce")
# Convert numeric columns
```

```python
for col in num_cols:
    df_final[col].fillna(df_final[col].median(), inplace=True)
# Missing numeric values
```

```python
df_final.isnull().sum()
```

```python
# # Age
# df_final["age"].fillna(df_final["age"].median(), inplace=True)

# # Sex
# df_final["sex"].fillna(df_final["sex"].mode()[0], inplace=True)

# # Lab values
# for col in ["TSH","T3","TT4","T4U","FTI"]:
#     df_final[col].fillna(df_final[col].median(), inplace=True)
```

```python
pd.options.mode.chained_assignment = None
df_final["age"] = df_final["age"].fillna(df_final["age"].median())

df_final["sex"] = df_final["sex"].fillna(df_final["sex"].mode()[0])

for col in ["TSH","T3","TT4","T4U","FTI"]:
    df_final[col] = df_final[col].fillna(df_final[col].median())
```

```python
df_final.isnull().sum()
```

```python
df_final.replace({"t":1, "f":0}, inplace=True)

num_cols = ["TSH","T3","TT4","T4U","FTI","age"]

for col in num_cols:
    df_final[col] = pd.to_numeric(df_final[col], errors="coerce")
    df_final[col] = df_final[col].fillna(df_final[col].median())
```

```python
def classify_thyroid(row):
    if row["TSH"] > 4 and row["FTI"] < 80:
        return "hypothyroid"
    elif row["TSH"] < 0.4 and row["FTI"] > 120:
        return "hyperthyroid"
    else:
        return "normal"

df_final["target"] = df_final.apply(classify_thyroid, axis=1)    
```

```python
df_final.to_csv("cleaned_thyroid_dataset.csv", index=False)
```
